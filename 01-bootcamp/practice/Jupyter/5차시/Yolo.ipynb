```python
import cv2
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os
```

```python
model = YOLO('yolov8n.pt')
```

```python
image_files = [f for f in os.listdir() if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

if not image_files:
    print("No image files found in the current directory.")
else:
    print("Available image files:", image_files)
```

```python
image_files = [f for f in os.listdir() if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

if not image_files:
    print("No image files found in the current directory.")
else:
    print("Available image files:", image_files)

    image_name = input("Enter the image filename (from the above list) : ").strip()

    if image_name not in image_files:
        print("Invalid filename! Please run the script again and enter a valid image name.")
    else:
        results = model(image_name)

        results[0].show()
        results[0].save('detected_image.jpg')

        detected_img = cv2.imread('detected_image.jpg')
        detected_img = cv2.cvtColor(detected_img, cv2.COLOR_BGR2RGB)

        plt.imshow(detected_img)
        plt.axis("off")
        plt.title("Detected Objects")
        plt.show()
```
