```python
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset("titanic")
titanic.to_csv('titanic.csv', index = False)
```

```python
titanic.isnull().sum()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['embarked'].value_counts()
```

```python
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic['embark_town'].value_counts()
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Southamton')
titanic['deck'].value_counts()
```

```python
titanic['deck'] = titanic['deck'].fillna('C')
titanic.isnull().sum()
```

```python
titanic.info()
```

```python
titanic.survived.value_counts()
```

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(1,2,figsize=(10,5))
titanic['survived'][titanic['sex'] == 'male'].value_counts().plot.pie(explode = [0, 0.1], autopct = '%1.1f%%', ax = ax[0], shadow=True)
titanic['survived'][titanic['sex'] == 'female'].value_counts().plot.pie(explode = [0, 0.1], autopct = '%1.1f%%', ax = ax[1], shadow=True)
ax[0].set_title('Survived (male)')
ax[1].set_title('Survived (female)')
plt.show()
```

```python
sns.countplot(x ='pclass', hue = 'survived', data = titanic)
plt.title('Pclass vs Survived')
plt.show()
```

```python
titanic_numeric_bool = titanic.select_dtypes(include=['int64', 'float64', 'bool'])
titanic_corr = titanic_numeric_bool.corr(method='pearson')
titanic_corr
```

```python
titanic_corr.to_csv('titanic_corr.csv', index = False)
```

```python
sns.catplot(x = 'pclass', y = 'survived', hue = 'sex', data = titanic, kind = 'point')
plt.show()
```

```python
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
titanic['age2'] = titanic['age'].apply(category_age)
titanic['sex'] = titanic['sex'].map({'male':1, 'female':0})
titanic['family'] = titanic['sibsp'] + titanic['parch']
titanic.to_csv('titanic3.csv', index = False)
heatmap_data = titanic[['survived', 'sex', 'age2', 'family', 'pclass', 'fare']]
colormap = plt.cm.RdBu
plt.figure(figsize=(10, 8))
sns.heatmap(
    heatmap_data.astype(float).corr(),
    annot=True,
    cmap=colormap,
    linewidths=0.2
)
plt.title("Titanic Survival Correlation Heatmap")
plt.show()
```
